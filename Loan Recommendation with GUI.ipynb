```python
#importing Libraries
from tkinter import *
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np


#This functions takes the user input in data frame fromat and predicts the output after cleaning
def classifier(af):
    
    #importing Dataset
    dataset = ("loan_stat542.csv")
    df = pd.read_csv(dataset)


    #Data Preprocessing

    #Droping Useless columns
    df.drop('id', axis=1, inplace=True)
    df.drop('earliest_cr_line', axis=1, inplace=True)
    df.drop('application_type', axis=1, inplace=True)

    #Removing Rows Containing Empty Values
    df = df.dropna(how='any')
    
    # loan_status   droping rows with default values and standardizing 
    df = df.drop(13887, axis=0)
    df = df.drop(34900, axis=0)
    le_loan_status = LabelEncoder()
    df['loan_status'] = le_loan_status.fit_transform(df['loan_status'])
    
    #Appending New row
    df= df.append(af,ignore_index=True)
    #print(df.tail(1))

    # Term   removing string "months"
    df.term = df.term.str.replace('months', '').astype(int)
   

    # Int_rate   converting float into int 
    df.int_rate = round(df.int_rate).astype(int)

    #Installment   convertig float into int
    df.installment = round(df.installment).astype(int)
    
    #grade  standardizing
    le_grade = LabelEncoder()
    df['grade'] = le_grade.fit_transform(df['grade'])

    #sub_grade  standardizing
    le_sub_grade = LabelEncoder()
    df['sub_grade'] = le_sub_grade.fit_transform(df['sub_grade'])

    #emp_title   
    le_emp_title = LabelEncoder()
    df['emp_title'] = le_emp_title.fit_transform(df['emp_title'])

    #emp_length   removing string part
    df.emp_length = df.emp_length.str.replace('years', '')
    df.emp_length = df.emp_length.str.replace('year', '')
    df.emp_length = df.emp_length.str.replace('<', '')
    df.emp_length = df.emp_length.str.replace('>', '')
    df.emp_length = df.emp_length.str.replace('+', '').astype(int)

    #home_ownership  
    le_home_ownership = LabelEncoder()
    df['home_ownership'] = le_home_ownership.fit_transform(df['home_ownership'])

    #verification_status    
    le_verification_status = LabelEncoder()
    df['verification_status'] = le_verification_status.fit_transform(df['verification_status'])

    #purpose    
    le_purpose = LabelEncoder()
    df['purpose'] = le_purpose.fit_transform(df['purpose'])

    #title    Standardization
    le_title = LabelEncoder()
    df['title'] = le_title.fit_transform(df['title'])


    #addr_state     Standardization
    le_addr_state = LabelEncoder()
    df['addr_state'] = le_addr_state.fit_transform(df['addr_state'])

    #dti  converting from float to int
    df.dti = round(df.dti).astype(int)


    #revol_util   Converting from float to int
    df.revol_util = round(df.revol_util).astype(float)
    
    #initial_list_status
    le_initial_list_status = LabelEncoder()
    df['initial_list_status'] = le_initial_list_status.fit_transform(df['initial_list_status'])


    #Splitting Data
    x = df[['loan_amnt', 'term', 'int_rate', 'installment', 'grade',
       'sub_grade', 'emp_title', 'emp_length', 'home_ownership', 'annual_inc',
       'verification_status', 'purpose', 'title', 'addr_state', 'dti',
       'fico_range_low', 'fico_range_high', 'open_acc',
       'pub_rec', 'revol_bal', 'revol_util', 'total_acc',
       'initial_list_status', 'mort_acc',
       'pub_rec_bankruptcies']]
    y=df['loan_status']

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.6, random_state=0)



    #creating classifer
    clf = LogisticRegression(solver ='liblinear')
    clf.fit(x_train, y_train)
    y_predict = clf.predict(x_test)
    
    #seprating the user input and sending the results to GUI function    
    res=df.tail(1)
    res.drop('loan_status', axis=1, inplace=True)
    r = clf.predict(res)
    return r





#GUI start here
root =Tk()
root.geometry("800x600")
root.maxsize(800, 600)
root.minsize(800,600)
root.title("Loan Recommendation System")


#This function takes the user input, convert it to a dataframe and passes the value to classifier
def test():
    a=[[l_amnt.get(),trm.get(),int_rate.get(),instalmnt.get(),grade.get(),sgrade.get(),empt.get(),empl.get(),homOner.get(),aincom.get(),vstatus.get(),purpose.get(),title.get(),state.get(),dti.get(),frl.get(),frh.get(),openacc.get(),prec.get(),rbal.get(),rutil.get(),tacc.get(),ils.get(),macc.get(),prb.get(),lstatus.get()]]    
    af=pd.DataFrame(a, columns= ['loan_amnt', 'term', 'int_rate', 'installment', 'grade','sub_grade', 'emp_title', 'emp_length', 'home_ownership', 'annual_inc','verification_status', 'purpose', 'title', 'addr_state', 'dti','fico_range_low', 'fico_range_high', 'open_acc','pub_rec', 'revol_bal', 'revol_util', 'total_acc','initial_list_status', 'mort_acc','pub_rec_bankruptcies','loan_status'])   
    result = classifier(af)
    print(result)
    
    #printing the results on the screen
    if(result==0):
        status.config(text="Not Approved")
    else:
        status.config(text="Loan Approved")
    
    
    

#Heading
Label(text="Loan Recommendation System",
fg="white", bg="purple", font="Arial 18 bold", padx=220, pady=10 ).place(x=0, y=0)



#Text for Entries

#Column One 
Label(text="Loan Amount:", font="Arial 11").place( x=90, y=80)
Label(text="Term:", font="Arial 11").place( x=90, y=110)
Label(text="Intrest Rate:", font="Arial 11").place( x=90, y=140)
Label(text="Installment:", font="Arial 11").place( x=90, y=170)
Label(text="Grade:", font="Arial 11").place( x=90, y=200)
Label(text="Sub Grade:", font="Arial 11").place( x=90, y=230)
Label(text="Emplloyee Title:", font="Arial 11").place( x=90, y=260)
Label(text="Employee Lenght:", font="Arial 11").place( x=90, y=290)
Label(text="Home Ownership:", font="Arial 11").place( x=90, y=320)
Label(text="Annual Income:", font="Arial 11").place( x=90, y=350)
Label(text="Verification Status:", font="Arial 11").place( x=90, y=380)
Label(text="Purpose:", font="Arial 11").place( x=90, y=410)
Label(text="Title:", font="Arial 11").place( x=90, y=440)

#Column Two
Label(text="State:", font="Arial 11").place( x=430, y=80)
Label(text="DTI:", font="Arial 11").place( x=430, y=110)
Label(text="Fico Range Low:", font="Arial 11").place( x=430, y=140)
Label(text="Fico Range High:", font="Arial 11").place( x=430, y=170)
Label(text="Open Accounts:", font="Arial 11").place( x=430, y=200)
Label(text="Public Rec:", font="Arial 11").place( x=430, y=230)
Label(text="Revol Balance:", font="Arial 11").place( x=430, y=260)
Label(text="Revol Util:", font="Arial 11").place( x=430, y=290)
Label(text="Total Accounts:", font="Arial 11").place( x=430, y=320)
Label(text="Initial List Status:", font="Arial 11").place( x=430, y=350)
#Label(text="Application Type: will be deleted", font="Arial 11").place( x=430, y=380)
Label(text="Mort Acc:", font="Arial 11").place( x=430, y=380)
Label(text="Pub Rec Bankruptcies:", font="Arial 11").place( x=430, y=410)



#Variable for storing User Data
l_amnt=IntVar()
trm=StringVar()
int_rate =DoubleVar()
instalmnt=IntVar()
grade=StringVar()
sgrade=StringVar()
empt=StringVar()
empl=StringVar()
homOner=StringVar()
aincom=IntVar()
vstatus=StringVar()
purpose=StringVar()
title=StringVar()
state=StringVar()
dti=DoubleVar()
frl=IntVar()
frh=IntVar()
openacc=IntVar()
prec=IntVar()
rbal=IntVar()
rutil=DoubleVar()
tacc=IntVar()
ils=StringVar()
macc=IntVar()
prb=IntVar()
lstatus=IntVar()




#choices = { 'Pizza','Lasagne','Fries','Fish','Potatoe'}
#Chocies
Term = { '36 months','60 months'}
Grade = {'A','B','C','D','E','F','G'}
Sgrade = {'C1', 'A3', 'D1', 'C3', 'B5', 'B4', 'C4', 'D5', 'B3', 'C5', 'D2',
       'C2', 'B2', 'A4', 'D3', 'B1', 'A5', 'E5', 'F1', 'E4', 'D4', 'E2',
       'E1', 'G2', 'G1', 'F3', 'F2', 'A1', 'A2', 'G3', 'E3', 'F4', 'G4',
       'F5', 'G5'}
Home_Ownership = {'RENT', 'MORTGAGE', 'OWN'}
Verification_Status = {'Source Verified', 'Not Verified', 'Verified'}
Purpose= {'debt_consolidation', 'credit_card', 'home_improvement', 'house','medical', 'other', 'car', 'major_purchase', 'moving','small_business', 'vacation', 'renewable_energy', 'wedding'}
list_status={ 'f', 'w'}
Title={'Debt consolidation', 'Credit card refinancing','Home improvement', 'Home buying', 'Medical expenses', 'Other','Car financing', 'Major purchase', 'Moving and relocation','Business', 'Vacation', 'Green loan', 'debt payoff','Consolidation', 'debt pay off', 'Engagement Ring Purchase','Finish line','Debt Clean up', 'Getting out of debt'}



#Entries
Entry( root, textvariable=l_amnt).place( x=230, y=80)
OptionMenu( root, trm, *Term).place(x=225, y=105)
Entry( root, textvariable=int_rate).place( x=230, y=140)
Entry( root, textvariable=instalmnt).place( x=230, y=170)
OptionMenu( root, grade, *Grade).place(x=225, y=195)
OptionMenu( root, sgrade, *Sgrade).place(x=225, y=225)
Entry( root, textvariable=empt).place( x=230, y=260)
Entry( root, textvariable=empl).place( x=230, y=290)
OptionMenu( root, homOner, *Home_Ownership).place(x=225, y=320)
Entry( root, textvariable=aincom).place( x=230, y=350)
OptionMenu( root, vstatus, *Verification_Status).place(x=225, y=375)
OptionMenu( root, purpose, *Purpose).place(x=225, y=405)
OptionMenu( root, title, *Title).place(x=225, y=435)

#second column
Entry( root, textvariable=state).place( x=600, y=80)
Entry( root, textvariable=dti).place( x=600, y=110 )
Entry( root, textvariable=frl).place( x=600, y=140)
Entry( root, textvariable=frh).place( x=600, y=170)
Entry( root, textvariable=openacc).place( x=600, y=200)
Entry( root, textvariable=prec).place( x=600, y=230)
Entry( root, textvariable=rbal).place( x=600, y=260)
Entry( root, textvariable=rutil).place( x=600, y=290)
Entry( root, textvariable=tacc).place( x=600, y=320)
OptionMenu( root, ils, *list_status).place(x=600, y=345)
Entry( root, textvariable=macc).place( x=600, y=380)
Entry( root, textvariable=prb).place( x=600, y=410)


#Button
btn= Button(root, fg='white', bg='purple', text='Enter', font="Arial 10 bold", padx=20, pady=3, command=test).place(x=90, y=500)

#Result
status=Label(root, text='', font="Arial 18 bold", fg="red")
status.place(x=510, y=500)


root.iconbitmap('logo.ico')



root.mainloop()
```
